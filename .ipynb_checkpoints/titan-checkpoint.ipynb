```python
import os
import pandas as pd


imu_data_path = "/Volumes/DDH/Pitts/oinauto/wk1/TITAN/titan/titan_data/dataset/imu_data"

# Recursively find all CSV files 
imu_files = []
for root, _, files in os.walk(imu_data_path):
    for file in files:
        if file.endswith(".csv"):  
            imu_files.append(os.path.join(root, file))

print(f"Total IMU data files found: {len(imu_files)}")  

# Read the first IMU data file if available 
if imu_files:
    imu_sample_file = imu_files[0]
    df_imu = pd.read_csv(imu_sample_file)

    print("Sample IMU data:")
    print(df_imu.head())

    print("\n Dataset Info:")
    print(df_imu.info())

    print("\n Statistical Summary:")
    print(df_imu.describe())
else:
    print("No IMU data files found!")
```

```python
print("Actual Column Names:", df_imu.columns.tolist())
```

```python
# Define proper column names based on the TITAN dataset documentation
imu_column_names = [
    "image_ts",      # Timestamp when image was captured
    "image_path",    # File path of the corresponding image
    "accel_ts",      # Timestamp of accelerometer measurement
    "accel",         # Accelerometer data (m/s^2)
    "gyro_ts",       # Timestamp of gyroscope measurement
    "ang_vel"        # Angular velocity (rad/s)
]

df_imu = pd.read_csv(imu_sample_file, header=None, names=imu_column_names)

print(df_imu.head())
```

```python
missing_values = df_imu.isnull().sum()
print(" Missing Values per Column:")
print(missing_values)

print("\n Data Types:")
print(df_imu.dtypes)
```

```python
import matplotlib.pyplot as plt

# Plot acceleration over time 
plt.figure(figsize=(12, 5))
plt.plot(df_imu["accel_ts"], df_imu["accel"], label="Acceleration (m/s²)")
plt.xlabel("Accelerometer Timestamp")  
plt.ylabel("Acceleration (m/s²)")  
plt.title("Acceleration Data Over Time")  
plt.legend()
plt.show()

# Plot gyroscope data over time 
plt.figure(figsize=(12, 5))
plt.plot(df_imu["gyro_ts"], df_imu["ang_vel"], label="Angular Velocity (rad/s)")
plt.xlabel("Gyroscope Timestamp")  
plt.ylabel("Angular Velocity (rad/s)")
plt.title("Gyroscope Data Over Time")  
plt.legend()
plt.show()
```
